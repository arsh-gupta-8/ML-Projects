```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

filename = "titanic_dataset.csv"
titanic_data = pd.read_csv(filename)
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
titanic_data.columns
```

```python
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 0, 'female': 1})

y = titanic_data['Survived']
X = titanic_data[['Pclass', 'Sex', 'Age', 'SibSp']]
survival_model = DecisionTreeClassifier(random_state=1)
survival_model.fit(X, y)
```

```python
print("Making predictions for the following 5 people:")
print(X.head())
print("The predictions are")
print(survival_model.predict(X.head()))
```
